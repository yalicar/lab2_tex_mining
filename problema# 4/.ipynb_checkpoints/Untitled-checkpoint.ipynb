```python
import cv2
import math
import numpy as np 
import matplotlib.pyplot as plt 
# impotar imagen
imagen = cv2.imread('imagen4.jpg')
im = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
imagen=im
```

```python
plt.imshow(imagen)
```

```python
# funciones para operar
def grafic(img,tit):
    im=img
    fig, axs = plt.subplots(1,2, figsize=(35, 10))
    fig.suptitle(tit,fontsize=35)
    color = ('b', 'g', 'r') 
    
    for i, col in enumerate(color): 
        histr = cv2.calcHist([im],[i], None,[256],[0, 256])
        axs[1].set_title('Histograma',fontsize=20)
        axs[1].plot(histr, color = col)
        axs[1].axis(xmin=0,xmax=256)
    axs[0].set_title('Imagen',fontsize=20)
    axs[0].imshow(im, origin='upper')
    plt.show()
```

```python
# transformacion Negativa
def negative_trans(img_p):
    img_bgr=img_p
    height, width, _ = img_bgr.shape
    final = np.zeros((height, width, 3))
    for i in range(0, height - 1):
        for j in range(0, width - 1):
            pixel = img_bgr[i, j] 
            pixel[0] = 255 - pixel[0]   
            pixel[1] = 255 - pixel[1]
            pixel[2] = 255 - pixel[2]
            final[i, j] = pixel
    return(final)

# Tranfortmacion Lineal
def linear_trans(img_p):
    image=img_p
    max_value= image.max()
    min_value= image.min()
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            image[i,j][0] = ((image[i,j][0] - min_value)*255)
            image[i,j][1] = ((image[i,j][1] - min_value)*255) / (max_value - min_value)
            image[i,j][2] = ((image[i,j][2] - min_value)*255) / (max_value)
    return(image)

# transformacion logaritmica
def log_trans(img_p):
    image=img_p
    c = 255 / (np.log(1 + np.max(image)))
    img = c * np.log(1 + image)
    img = np.array(img, dtype = np.uint8)  
    return(img)

# ransfrmacion exponencial
def exp_trans(img_p, gamma):
    image=img_p
    img = np.arrayimage(255*(image / 255) ** gamma, dtype = 'uint8')
    return (img)
```

```python
neg=negative_trans(imagen)
lin=linear_trans(imagen)
log=log_trans(imagen)
exp=exp_trans(imagen,5)
```

```python
grafic(imagen,"imagen Original")
```

```python

grafic(neg,"imagen negativa")
```

```python
grafic(lin,"imagen lineal")
```

```python
grafic(log,"Grafica logaritmica")
```

```python
grafic(exp,"Grafica exponencial")
```
